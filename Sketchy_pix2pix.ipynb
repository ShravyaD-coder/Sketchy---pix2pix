```python
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim
from skimage.io import imread
from tqdm import tqdm
import re
import numpy as np

image_folder = "/content/content/pytorch-CycleGAN-and-pix2pix/results/sketchy3/test_latest/images"

all_files = os.listdir(image_folder)


fake_B_files = sorted([f for f in all_files if f.endswith("_fake_B.png")])
real_B_files = sorted([f for f in all_files if f.endswith("_real_B.png")])



assert len(fake_B_files) == len(real_B_files), "Mismatch in number of fake and real images!"

psnr_scores = []
ssim_scores = []


for fake_file, real_file in tqdm(zip(fake_B_files, real_B_files), total=len(fake_B_files)):

    fake_img = imread(os.path.join(image_folder, fake_file))
    real_img = imread(os.path.join(image_folder, real_file))


    assert fake_img.shape == real_img.shape, f"Shape mismatch: {fake_file} and {real_file}"


    psnr_value = psnr(real_img, fake_img, data_range=real_img.max() - real_img.min())
    ssim_value = ssim(real_img, fake_img, channel_axis=-1, data_range=real_img.max() - real_img.min())

    psnr_scores.append(psnr_value)
    ssim_scores.append(ssim_value)


avg_psnr = np.mean(psnr_scores)
avg_ssim = np.mean(ssim_scores)


print(f"\nAverage PSNR: {avg_psnr:.2f}dB")
print(f"Average SSIM: {avg_ssim:.4f}")
```

```python
image_folder = "/content/content/pytorch-CycleGAN-and-pix2pix/results/sketchy3/test_latest/images"

all_files = os.listdir(image_folder)


fake_B_files = sorted([f for f in all_files if f.endswith("_fake_B.png")])
real_B_files = sorted([f for f in all_files if f.endswith("_real_B.png")])


assert len(fake_B_files) == len(real_B_files), "Mismatch in number of fake and real images!"


psnr_scores = {'even': [], 'odd': []}
ssim_scores = {'even': [], 'odd': []}


def extract_leading_number(filename):
    match = re.match(r'^(\d+)_fake_B\.png$', filename)
    return int(match.group(1)) if match else None


for fake_file, real_file in tqdm(zip(fake_B_files, real_B_files), total=len(fake_B_files)):
    image_id = extract_leading_number(fake_file)
    if image_id is None:
        continue

    category = 'even' if image_id % 2 == 0 else 'odd'


    fake_img = imread(os.path.join(image_folder, fake_file))
    real_img = imread(os.path.join(image_folder, real_file))

    # Check for shape mismatch
    assert fake_img.shape == real_img.shape, f"Shape mismatch: {fake_file} and {real_file}"


    psnr_value = psnr(real_img, fake_img, data_range=real_img.max() - real_img.min())
    ssim_value = ssim(real_img, fake_img, channel_axis=-1, data_range=real_img.max() - real_img.min())


    psnr_scores[category].append(psnr_value)
    ssim_scores[category].append(ssim_value)


for category in ['even', 'odd']:
    avg_psnr = np.mean(psnr_scores[category]) if psnr_scores[category] else 0
    avg_ssim = np.mean(ssim_scores[category]) if ssim_scores[category] else 0
    print(f"\n[{category.upper()} IMAGES]")
    print(f"Average PSNR: {avg_psnr:.2f} dB")
    print(f"Average SSIM: {avg_ssim:.4f}")
```
